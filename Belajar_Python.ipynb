## Quiz

Lets say our expense (pengeluaran) for every month are listed below: 

- January: 2200
- February: 2350
- March: 2600
- April: 2130
- May: 2190

Create list to store these monthly expenses: 

- In Feb, how many dollars you spent extra compare to January?
- Find out your total expenses in first quarter (first three months) of the year
- Find out if you spent exactly 2000 dollars in any month
- June month is just finished, and your expenses is 1980 dollar, add it to our month expenses list
- Kamu dapat kembalian dari barang yang kamu beli di bulan April dan dapat refund sebesar 200 dollar. make correction to your monthly expenses list based on this.

### Jawaban

```python
expenses = [2200, 2350, 2600, 2130, 2190]
```

```python
print(expenses[0] - expenses[1]) # 1. Spent extra compared to January
```

```python
print(expenses[0] + expenses[1] + expenses[2]) # 2. Total expenses in first quarter
```

```python
2000 in expenses # 3. Find out if you spent exactly 2000 dollars in any month
```

```python
expenses.append(1980) # 4. Add June expense to Monthly expenses
```

```python
expenses[3] + 200
```

# Tuples

Tuple itu sama persis dengan list/array, perbedaannya adalah, Tuple bersifat immutable, yang mana setelah kita membuat Tuple, kita tidak bisa mengganti value di dalamnya, jadinya permanen.

```python
x = (8, 3, 7, 2, 5, 4, 6, 9, 9)
```

```python
x[0: 3]
```

```python
x.sort() # ini akan menghasilkan error, karena, Tuple bersifat immutable (permanen)
```

```python
x.count(9) # perintah count() tidak akan menghasilkan error, karena tidak mengubah isi value dari tuple tersebut
```

```python
y = ('Indra', 'Damar', 'Jati')
```

```python
z = x + y
```

```python
z # ini bisa terjadi, karena, ini tidak mengubah value tuple asal, yaitu x dan y
```
